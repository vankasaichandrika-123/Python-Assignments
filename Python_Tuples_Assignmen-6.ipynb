## Part B — Coding Questions (No for-loop)

### 1. Create an empty tuple.

```python
empty_tuple = ()
print("Empty tuple:", empty_tuple)
print("Type:", type(empty_tuple))
```

### 2. Create a tuple with 5 integers.

```python
t = (10, 20, 30, 40, 50)
print("Tuple of 5 integers:", t)
```

### 3. Access 3rd element of a tuple.

```python
t = (10, 20, 30, 40, 50)
print("3rd element:", t[2])
```

### 4. Access last element without negative index.

```python
t = (10, 20, 30, 40, 50)
print("Last element:", t[len(t) - 1])
```

### 5. Slice first three elements.

```python
t = (10, 20, 30, 40, 50)
print("First three elements:", t[:3])
```

### 6. Slice every alternate value.

```python
t = (10, 20, 30, 40, 50, 60)
print("Every alternate value:", t[::2])
```

### 7. Reverse a tuple using slicing.

```python
t = (10, 20, 30, 40, 50)
print("Reversed tuple:", t[::-1])
```

### 8. Concatenate two tuples.

```python
t1 = (1, 2, 3)
t2 = (4, 5, 6)
combined = t1 + t2
print("Concatenated tuple:", combined)
```

### 9. Replicate tuple (1,2) twice.

```python
t = (1, 2)
replicated = t * 2
print("Replicated tuple:", replicated)
```

### 10. Find length of a tuple.

```python
t = (10, 20, 30, 40)
print("Length of tuple:", len(t))
```

### 11. Convert a list to a tuple.

```python
lst = [1, 2, 3, 4]
t = tuple(lst)
print("Converted tuple:", t)
```

### 12. Convert a string "python" to tuple of characters.

```python
s = "python"
t = tuple(s)
print("Tuple of characters:", t)
```

### 13. Check if value exists in tuple.

```python
t = (10, 20, 30, 40)
value = 30
print("Value exists in tuple:", value in t)
```

### 14. Find maximum element in tuple.

```python
t = (12, 45, 7, 89, 23)
print("Maximum element:", max(t))
```

### 15. Find minimum element in tuple.

```python
t = (12, 45, 7, 89, 23)
print("Minimum element:", min(t))
```

### 16. Count how many times 2 appears in (1,2,2,3).

```python
t = (1, 2, 2, 3)
print("Count of 2:", t.count(2))
```

### 17. Find index of an element.

```python
t = (10, 20, 30, 40)
print("Index of 30:", t.index(30))
```

### 18. Perform tuple packing.

```python
# Tuple packing: assigning multiple values to a single tuple variable
packed = 10, 20, 30
print("Packed tuple:", packed)
print("Type:", type(packed))
```

### 19. Perform tuple unpacking into variables.

```python
t = (10, 20, 30)
a, b, c = t
print("a =", a)
print("b =", b)
print("c =", c)
```

### 20. Convert tuple to string without loops.

```python
t = ('p', 'y', 't', 'h', 'o', 'n')
s = "".join(t)
print("Converted string:", s)
```
