# Python String Programs (20 Questions with Answers)

A collection of basic-to-intermediate Python string manipulation exercises, solved using indexing, slicing, and built-in string methods.

### 1. Take a string from user and print its length.

```
s = input("Enter a string: ")
print("Length of string:", len(s))
```

### 2. Input a string and print first and last character.

```
s = input("Enter a string: ")
print("First character:", s[0])
print("Last character:", s[-1])
```

### 3. Input a string and print character at index 2 and -2.

```
s = input("Enter a string: ")
print("Character at index 2:", s[2])
print("Character at index -2:", s[-2])
```

### 4. Input a string and print the middle character(s).

```
s = input("Enter a string: ")
n = len(s)
if n % 2 == 0:
    mid = s[n//2 - 1 : n//2 + 1]
else:
    mid = s[n//2]
print("Middle character(s):", mid)
```

### 5. Reverse a string using slicing.

```
s = input("Enter a string: ")
print("Reversed string:", s[::-1])
```

### 6. Extract first 5 characters and last 5 characters.

```
s = input("Enter a string: ")
print("First 5 characters:", s[:5])
print("Last 5 characters:", s[-5:])
```

### 7. Print every second character from a string.

```
s = input("Enter a string: ")
print("Every second character:", s[::2])
```

### 8. Print characters from 2nd index to 7th index.

```
s = input("Enter a string: ")
print("Characters from index 2 to 7:", s[2:8])
```

### 9. Remove leading and trailing spaces from input string.

```
s = input("Enter a string with spaces: ")
print("Stripped string:", s.strip())
```

### 10. Convert string to uppercase and lowercase.

```
s = input("Enter a string: ")
print("Uppercase:", s.upper())
print("Lowercase:", s.lower())
```

### 11. Replace all spaces in a string with -.

```
s = input("Enter a string: ")
print("Replaced string:", s.replace(" ", "-"))
```

### 12. Check if a string starts with "A" or ends with "Z".

```
s = input("Enter a string: ")
print("Starts with 'A':", s.startswith("A"))
print("Ends with 'Z':", s.endswith("Z"))
```

### 13. Count occurrences of a character in string using count().

```
s = input("Enter a string: ")
ch = input("Enter character to count: ")
print(f"'{ch}' occurs", s.count(ch), "times")
```

### 14. Find position of substring in given string.

```
s = input("Enter a string: ")
sub = input("Enter substring to find: ")
print("Position of substring:", s.find(sub))
```

### 15. Split a string into words and print each word separately (no loop — use index access methods).

```
s = input("Enter a sentence: ")
words = s.split()
print("Word 1:", words[0])
print("Word 2:", words[1] if len(words) > 1 else "N/A")
print("Word 3:", words[2] if len(words) > 2 else "N/A")
print("All words:", words)
```

### 16. Check if input is numeric or alphabetic.

```
s = input("Enter a string: ")
print("Is numeric:", s.isnumeric())
print("Is alphabetic:", s.isalpha())
```

### 17. Swap first and last characters of string using slicing.

```
s = input("Enter a string: ")
swapped = s[-1] + s[1:-1] + s[0]
print("Swapped string:", swapped)
```

### 18. Check if two strings are equal (case-insensitive).

```
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
print("Strings are equal (case-insensitive):", s1.lower() == s2.lower())
```

### 19. Create a substring by skipping characters: s[1:len(s):2].

```
s = input("Enter a string: ")
print("Substring (skipping characters):", s[1:len(s):2])
```

### 20. Join a list of strings into one string separated by commas.

```
words = ["Python", "Java", "SQL", "AWS"]
joined = ",".join(words)
print("Joined string:", joined)
```
