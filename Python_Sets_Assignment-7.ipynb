**Notes on tricky ones:**
- **Q8:** `s = {}` creates an empty **dict**, not a set — to create an empty set you must use `set()`.
- **Q11 vs Q29 (remove vs discard):** `remove()` raises a `KeyError` if the element doesn't exist; `discard()` does **not** raise an error — it silently does nothing if the element is absent.
- **Q23:** A list is mutable and unhashable, so it cannot be an element of a set (`{1,[2,3]}` raises a `TypeError`).


## Part B — Coding Questions (20)

### 1. Create a set with five numbers.

```python
s = {10, 20, 30, 40, 50}
print("Set with five numbers:", s)
```

### 2. Add an element to a set.

```python
s = {10, 20, 30}
s.add(40)
print("Set after adding element:", s)
```

### 3. Remove an element from a set if present.

```python
s = {10, 20, 30, 40}
s.discard(20)   # safe removal, no error if absent
print("Set after removing 20:", s)
```

### 4. Convert a list into a set.

```python
lst = [1, 2, 2, 3, 4, 4]
s = set(lst)
print("Set from list:", s)
```

### 5. Check if an element exists in a set.

```python
s = {10, 20, 30, 40}
print("30 exists in set:", 30 in s)
```

### 6. Find union of two sets.

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print("Union:", s1 | s2)
```

### 7. Find intersection of two sets.

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print("Intersection:", s1 & s2)
```

### 8. Find difference between two sets.

```python
s1 = {1, 2, 3}
s2 = {2, 4}
print("Difference (s1 - s2):", s1 - s2)
```

### 9. Find symmetric difference between two sets.

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print("Symmetric Difference:", s1 ^ s2)
```

### 10. Remove all elements from a set.

```python
s = {10, 20, 30}
s.clear()
print("Set after clearing:", s)
```

### 11. Create a frozen set from a list.

```python
lst = [1, 2, 3, 4]
fs = frozenset(lst)
print("Frozen set:", fs)
print("Type:", type(fs))
```

### 12. Convert a string into a set of characters.

```python
s = "python"
char_set = set(s)
print("Set of characters:", char_set)
```

### 13. Check if one set is subset of another.

```python
s1 = {1, 2}
s2 = {1, 2, 3, 4}
print("s1 is subset of s2:", s1.issubset(s2))
```

### 14. Check if one set is superset of another.

```python
s1 = {1, 2, 3, 4}
s2 = {1, 2}
print("s1 is superset of s2:", s1.issuperset(s2))
```

### 15. Check if two sets are disjoint.

```python
s1 = {1, 2, 3}
s2 = {4, 5, 6}
print("Sets are disjoint:", s1.isdisjoint(s2))
```

### 16. Find length of a set.

```python
s = {10, 20, 30, 40}
print("Length of set:", len(s))
```

### 17. Pop an element from set and display result.

```python
s = {10, 20, 30, 40}
removed = s.pop()
print("Popped element:", removed)
print("Set after pop:", s)
```

### 18. Copy a set into another set variable.

```python
s1 = {10, 20, 30}
s2 = s1.copy()
print("Original set:", s1)
print("Copied set:", s2)
```

### 19. Create a set from a tuple.

```python
t = (1, 2, 3, 2, 1)
s = set(t)
print("Set from tuple:", s)
```

### 20. Create two sets and show elements unique to each (symmetric diff).

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
unique_elements = s1 ^ s2
print("Set 1:", s1)
print("Set 2:", s2)
print("Elements unique to each set (Symmetric Difference):", unique_elements)
```
