## Python Operators Assignment

Solutions using arithmetic, comparison, logical, assignment, and bitwise operators.

### 1. Input two numbers and print their Sum, Difference, and Product.

```
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
```

### 2. Check whether a given number is even or odd using the modulus (%) operator.

```
num = int(input("Enter a number: "))

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

### 3. Accept an integer and print its Square and Cube (using ** operator).

```
num = int(input("Enter a number: "))

print("Square:", num ** 2)
print("Cube:", num ** 3)
```

### 4. Compare two numbers — check equality and which is greater (using comparison operators).

```
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Are equal:", a == b)

if a > b:
    print("First number is greater")
elif b > a:
    print("Second number is greater")
else:
    print("Both numbers are equal")
```

### 5. Check if a number lies between 10 and 50 (inclusive) using logical operators.

```
num = int(input("Enter a number: "))

if num >= 10 and num <= 50:
    print(num, "lies between 10 and 50 (inclusive)")
else:
    print(num, "does not lie between 10 and 50")
```

### 6. Swap values of two variables without using a third variable (using assignment operators).

```
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

print("Before swap -> a:", a, ", b:", b)

a, b = b, a   # Pythonic swap (no third variable used)

print("After swap  -> a:", a, ", b:", b)
```

### 7. Take a string from the user and check if it contains the character 'a'.

```
s = input("Enter a string: ")

if 'a' in s:
    print("The string contains the character 'a'")
else:
    print("The string does not contain the character 'a'")
```

### 8. Check whether a given number is both divisible by 3 and 5.

```
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print(num, "is divisible by both 3 and 5")
else:
    print(num, "is not divisible by both 3 and 5")
```

### 9. Accept three numbers and print the largest using comparison and logical operators.

```
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("Largest number is:", largest)
```

### 10. Calculate the area of a circle (area = 3.14 * r ** 2).

```
r = float(input("Enter radius: "))

area = 3.14 * r ** 2
print("Area of circle:", area)
```

### 11. Check if a number is positive, negative, or zero.

```
num = float(input("Enter a number: "))

if num > 0:
    print(num, "is Positive")
elif num < 0:
    print(num, "is Negative")
else:
    print(num, "is Zero")
```

### 12. Toggle the case of a given alphabet using ASCII operators.

```
ch = input("Enter a single alphabet: ")

if ch.isupper():
    toggled = chr(ord(ch) + 32)   # Uppercase -> lowercase
elif ch.islower():
    toggled = chr(ord(ch) - 32)   # Lowercase -> uppercase
else:
    toggled = ch

print("Toggled character:", toggled)
```

### 13. Input a number and check if it is a multiple of either 4 or 6.

```
num = int(input("Enter a number: "))

if num % 4 == 0 or num % 6 == 0:
    print(num, "is a multiple of either 4 or 6")
else:
    print(num, "is not a multiple of 4 or 6")
```

### 14. Perform bitwise operations (AND, OR) between two numbers.

```
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("a & b (Bitwise AND):", a & b)
print("a | b (Bitwise OR):", a | b)
```
