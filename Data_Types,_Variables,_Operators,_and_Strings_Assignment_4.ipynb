```python
#Take a string input and count how many vowels and consonants it has.
s = input().lower()

vowels = (s.count('a') + s.count('e') + s.count('i') +
          s.count('o') + s.count('u'))

letters = len(s.replace(" ", ""))
consonants = letters - vowels

print("Vowels =", vowels)
print("Consonants =", consonants)
```

**Part C: Logical and Conceptual Questions**


```python
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)
```

```python
# Mutable
lst = [1, 2, 3]
lst.append(4)   # same object modified
print(lst)      # [1, 2, 3, 4]

# Immutable
s = "hello"
s = s + " world"   # new string object created, old one discarded
print(s)            # "hello world"
```

```python
a = "hello"
b = "hello"
print(a == b)   # True  (same value)
print(a is b)   # True  (Python interns small string literals, so same memory)

c = "hello world!"
d = "hello world!"
print(c == d)   # True
print(c is d)   # Often False (longer/dynamic strings usually aren't interned)
```

```python
import copy

original = [[1, 2], [3, 4]]

shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[0][0] = 99
print(shallow)  # [[99, 2], [3, 4]]  -> affected (shares inner list)
print(deep)     # [[1, 2], [3, 4]]   -> unaffected (fully independent)
```

```python
a = "100"
b = int(a)        # explicit casting: string -> int
print(b + 50)      # 150

x = 5
y = 2.0
z = x + y           # implicit casting: int automatically becomes float
print(z)             # 7.0
```

```python
a, b = 10, 5
print(a + b)        # Arithmetic -> 15
print(a > b)         # Relational -> True
print(a > b and b > 0)  # Logical    -> True
```

```python
x = 10        # x is an int
print(type(x))  # <class 'int'>

x = "hello"   # same variable, now a string
print(type(x))  # <class 'str'>
```

```python
a = 100          # 'a' points to int object 100
print(id(a))

a = 200          # 'a' now points to a new object 200
print(id(a))      # different memory address; old object 100 may get garbage-collected if unused elsewhere
```
