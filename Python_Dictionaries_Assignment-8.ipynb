**Notes on tricky ones:**
- **Q3:** Dictionary keys must be **immutable and hashable** — so tuples are valid keys, but lists and dicts are not.
- **Q23:** Dictionaries don't support direct slicing. To slice, you must first convert keys/values/items to a list (e.g., `list(d.keys())[:2]`).
- **Q26:** `max()` on a dictionary by default iterates over its **keys**, not values — so it compares `'a'`, `'b'`, `'c'` as strings and returns `'c'` (the alphabetically largest key).


## Section B — Coding Questions (20)

### 1. Create a dictionary with 3 key-value pairs and print it.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Dictionary:", student)
```

### 2. Access and print the value of a specific key from a dictionary.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Name:", student["name"])
```

### 3. Use get() to retrieve a value safely without causing an error.

```python
student = {"name": "Saichandrika", "age": 25}
print("Role (using get, key may not exist):", student.get("role", "Not Specified"))
```

### 4. Merge two dictionaries into a new one using update().

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged = dict1.copy()
merged.update(dict2)
print("Merged dictionary:", merged)
```

### 5. Remove a specific key using pop() and print the updated dictionary.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
student.pop("age")
print("Dictionary after removing 'age':", student)
```

### 6. Check if a key exists in a dictionary using in operator.

```python
student = {"name": "Saichandrika", "age": 25}
print("'name' exists:", "name" in student)
print("'salary' exists:", "salary" in student)
```

### 7. Retrieve all the keys from a dictionary using keys() method.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Keys:", student.keys())
```

### 8. Retrieve all the values from a dictionary using values() method.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Values:", student.values())
```

### 9. Get all key-value pairs from a dictionary using items().

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Items:", student.items())
```

### 10. Create a copy of a dictionary and verify both are independent.

```python
original = {"a": 1, "b": 2}
copy_dict = original.copy()

copy_dict["c"] = 3   # modify the copy

print("Original dictionary:", original)
print("Copied dictionary:", copy_dict)
print("Are they independent:", original is not copy_dict)
```

### 11. Use setdefault() to insert a key only if it doesn't exist.

```python
student = {"name": "Saichandrika", "age": 25}
student.setdefault("role", "Developer")   # 'role' not present, gets added
student.setdefault("name", "Unknown")     # 'name' already exists, unchanged
print("Dictionary after setdefault:", student)
```

### 12. Convert a list of tuples [("a",1),("b",2)] into a dictionary.

```python
list_of_tuples = [("a", 1), ("b", 2)]
result = dict(list_of_tuples)
print("Dictionary from list of tuples:", result)
```

### 13. Convert two lists ["a","b"] and [1,2] into a dictionary.

```python
keys = ["a", "b"]
values = [1, 2]
result = dict(zip(keys, values))
print("Dictionary from two lists:", result)
```

### 14. Find the length of a dictionary using len().

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer"}
print("Length of dictionary:", len(student))
```

### 15. Delete all items from a dictionary using clear().

```python
student = {"name": "Saichandrika", "age": 25}
student.clear()
print("Dictionary after clear():", student)
```

### 16. Extract keys as a list and slice the first two.

```python
student = {"name": "Saichandrika", "age": 25, "role": "Java Developer", "city": "Hyderabad"}
keys_list = list(student.keys())
print("All keys:", keys_list)
print("First two keys:", keys_list[:2])
```

### 17. Sort the dictionary by keys and display it.

```python
student = {"role": "Java Developer", "age": 25, "name": "Saichandrika"}
sorted_dict = dict(sorted(student.items()))
print("Dictionary sorted by keys:", sorted_dict)
```

### 18. Access a nested dictionary value, e.g., {"emp":{"name":"John","age":30}}.

```python
data = {"emp": {"name": "John", "age": 30}}
print("Employee name:", data["emp"]["name"])
print("Employee age:", data["emp"]["age"])
```

### 19. Use dictionary comprehension (without loops) to create a dictionary from a string using dict(zip()).

```python
s = "abcd"
result = dict(zip(s, range(1, len(s) + 1)))
print("Dictionary from string:", result)
```

### 20. Find the maximum and minimum value in a dictionary using built-in functions.

```python
marks = {"Math": 85, "Science": 92, "English": 78}

max_value = max(marks.values())
min_value = min(marks.values())

print("Maximum value:", max_value)
print("Minimum value:", min_value)
```
