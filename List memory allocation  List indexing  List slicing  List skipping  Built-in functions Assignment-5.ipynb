# VIHARATECH CODING ASSIGNMENTS
## Python Lists — Coding Questions (No Loops)

Solved using indexing, slicing, and built-in list functions only.

### 1. Access the 3rd element of a list.

```
lst = [10, 20, 30, 40, 50]
print("3rd element:", lst[2])
```

### 2. Print the last element without using negative index.

```
lst = [10, 20, 30, 40, 50]
print("Last element:", lst[len(lst) - 1])
```

### 3. Slice a list to get first three items.

```
lst = [10, 20, 30, 40, 50]
print("First three items:", lst[:3])
```

### 4. Slice to get every second element.

```
lst = [10, 20, 30, 40, 50, 60]
print("Every second element:", lst[::2])
```

### 5. Reverse a list using slicing.

```
lst = [10, 20, 30, 40, 50]
print("Reversed list:", lst[::-1])
```

### 6. Create a shallow copy of a list.

```
lst = [10, 20, 30]
shallow_copy = lst.copy()   # or lst[:]
print("Original list:", lst)
print("Shallow copy:", shallow_copy)
```

### 7. Find length of a list.

```
lst = [10, 20, 30, 40]
print("Length of list:", len(lst))
```

### 8. Find max value in a list.

```
lst = [12, 45, 7, 89, 23]
print("Maximum value:", max(lst))
```

### 9. Find min value in a list.

```
lst = [12, 45, 7, 89, 23]
print("Minimum value:", min(lst))
```

### 10. Remove the last element from a list.

```
lst = [10, 20, 30, 40]
lst.pop()
print("List after removing last element:", lst)
```

### 11. Add an item to end of list.

```
lst = [10, 20, 30]
lst.append(40)
print("List after appending:", lst)
```

### 12. Insert element at index 1.

```
lst = [10, 20, 30]
lst.insert(1, 99)
print("List after insertion:", lst)
```

### 13. Convert string to list of characters.

```
s = "python"
char_list = list(s)
print("List of characters:", char_list)
```

### 14. Convert string "a b c" to ['a','b','c'].

```
s = "a b c"
result = s.split()
print("Converted list:", result)
```

### 15. Concatenate two lists.

```
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print("Concatenated list:", combined)
```

### 16. Check if value exists in list.

```
lst = [10, 20, 30, 40]
value = 30
print("Value exists in list:", value in lst)
```

### 17. Create a list from 1-5 using list constructor.

```
lst = list(range(1, 6))
print("List from 1 to 5:", lst)
```

### 18. Remove an item by value.

```
lst = [10, 20, 30, 40]
lst.remove(30)
print("List after removing value 30:", lst)
```

### 19. Clear all elements of a list.

```
lst = [10, 20, 30, 40]
lst.clear()
print("List after clearing:", lst)
```

### 20. Sort a list without modifying original list.

```
lst = [40, 10, 30, 20]
sorted_list = sorted(lst)
print("Original list:", lst)
print("Sorted list (new):", sorted_list)
```
